TO DOWNLOAD IMAGE FROM URLs.

```python
import os
import random
import requests

# Function to download an image from a URL
def download_image(url, save_path):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(save_path, "wb") as file:
            for chunk in response.iter_content(1024):
                file.write(chunk)
        print(f"Downloaded: {save_path}")
    else:
        print(f"Failed to download: {url}")

# Folder path to save the images
save_dir = "/content/original_images"

# Create the save directory if it doesn't exist
os.makedirs(save_dir, exist_ok=True)

# Number of images to download
num_images = 100

# Random image URLs
base_url = "https://source.unsplash.com/random"
image_urls = [f"{base_url}/{random.randint(800, 1200)}x{random.randint(800, 1200)}" for _ in range(num_images)]

# Download each image
for i, url in enumerate(image_urls):
    save_path = os.path.join(save_dir, f"image{i+1}.jpg")
    download_image(url, save_path)

print("Image download complete.")
```

```python
import os
import random
import requests
import cv2

# Function to download an image from a URL
def download_image(url, save_path):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(save_path, "wb") as file:
            for chunk in response.iter_content(1024):
                file.write(chunk)
        print(f"Downloaded: {save_path}")
    else:
        print(f"Failed to download: {url}")

# Function to convert image to grayscale
def convert_to_gray(image_path, save_path):
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(save_path, gray)
    print(f"Converted to grayscale: {image_path}")

# Folder paths
save_dir = "/content/original_images"
gray_save_dir = "/content/gray_images"

# Create the save directories if they don't exist
os.makedirs(gray_save_dir, exist_ok=True)
i=0
# Convert to grayscale
for i in range(101):
  image_path = os.path.join(save_dir, f"image{i+1}.jpg")
  gray_save_path = os.path.join(gray_save_dir, f"gray_image{i+1}.jpg")
  convert_to_gray(image_path, gray_save_path)

print("Image download and conversion complete.")
```
